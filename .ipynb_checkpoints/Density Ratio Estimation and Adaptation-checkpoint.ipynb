### 0. Covariate Shift Adaptation on 1-dim Data

```python
# make some data which has a clear covariate shift
n_samples_Z = 1000
n_samples_X = 100
x = 11*np.random.random(n_samples_Z)- 6.0
y = x**2 + 10*np.random.random(n_samples_Z) - 5
Z = np.c_[x, y]
    
x = 2*np.random.random(n_samples_X) - 6.0
y = x**2 + 10*np.random.random(n_samples_X) - 5
X = np.c_[x, y]
```

```python
# plot the data
plt.figure(figsize=(12,8))
plt.scatter(Z[:,0], Z[:,1], marker='o', s=4, c='b', label='Z')
plt.scatter(X[:,0], X[:,1], marker='o', s=4, c='r', label='X')
plt.legend()
```

```python
X = pd.DataFrame(X)
Z = pd.DataFrame(Z)
X['is_z'] = 0 # 0 means test set
Z['is_z'] = 1 # 1 means training set
XZ = pd.concat([X, Z], ignore_index=True, axis=0)

labels = XZ['is_z'].values
XZ = XZ.drop('is_z', axis=1).values
X, Z = X.values, Z.values
```

```python
# can use a non-linear learner, but make sure to restrict how 
# much it can learn or it will discriminate too well.
clf_RF = RFC(max_depth=2)
# because we can see a learn divide in the above data we 
# could simply use logistic regression here.
clf_LR = LR(solver='liblinear')

predictions_LR = np.zeros(labels.shape)
predictions_RF = np.zeros(labels.shape)
skf = SKF(n_splits=20, shuffle=True, random_state=1234)
for fold, (train_idx, test_idx) in enumerate(skf.split(XZ, labels)):
    #print 'Training discriminator model for fold {}'.format(fold)
    X_train, X_test = XZ[train_idx], XZ[test_idx]
    y_train, y_test = labels[train_idx], labels[test_idx]
        
    clf_LR.fit(X_train, y_train)
    clf_RF.fit(X_train, y_train)
    probs = clf_LR.predict_proba(X_test)[:, 1]
    predictions_LR[test_idx] = probs
    probs = clf_RF.predict_proba(X_test)[:, 1]
    predictions_RF[test_idx] = probs
```

```python
# first, isolate the training part (recall we joined them above)
predictions_Z = predictions_LR[len(X):]
weights = (1./predictions_Z) - 1. 
weights /= np.mean(weights) # we do this to re-normalize the computed log-loss
plt.figure(figsize = (12,16))
ax = plt.subplot(211)
plt.xlabel('Computed sample weight')
plt.ylabel('# Samples')
plt.title("Logistic Regression")
sns.distplot(weights, kde=False)

ax = plt.subplot(212)
Zsize = 0.1 + weights*15
plt.scatter(Z[:,0], Z[:,1], marker='o', s=Zsize, c='b', label='Z')
plt.scatter(X[:,0], X[:,1], marker='o', s=4, c='r', label='X')
plt.legend()
```

```python
predictions_Z = predictions_RF[len(X):]
weights = (1./predictions_Z) - 1. 
weights /= np.mean(weights)
plt.figure(figsize = (12,16))
ax = plt.subplot(211)
plt.xlabel('Computed sample weight')
plt.ylabel('# Samples')
plt.title("Random Foreset")
sns.distplot(weights, kde=False)

ax = plt.subplot(212)
Zsize = 0.1 + weights*15
plt.scatter(Z[:,0], Z[:,1], marker='o', s=Zsize, c='b', label='Z')
plt.scatter(X[:,0], X[:,1], marker='o', s=4, c='r', label='X')
plt.legend()
```

### 1.Visualization of 2 Variables Toy Example


```python
# First generate 2 highly correlated variables
np.random.seed(123)
size, corr = 1000, 0.99
mean = [0, 0]
cov = [[1, corr], [corr, 1]]
X = np.random.multivariate_normal(mean, cov, size)
a, b = X.T
print('Original Correlation Matrix')
print(np.corrcoef(X, rowvar=False))

# Then perform column-wise random sampling
new_size = 5000
a_new = [a[np.random.randint(0, size)] for i in range(new_size)]
b_new = [b[np.random.randint(0, size)] for i in range(new_size)]
X_new = np.vstack((a_new, b_new)).T
print('Ramdomize Correlation Matrix')
print(np.corrcoef(X_new, rowvar=False))
```

```python
area = [50]*size
fig = plt.figure(figsize=(12,12))
ax = plt.subplot()
ax.scatter(a, b, s=area, alpha=0.5)
ax.scatter(a_new, b_new, s=area, c='green', alpha=0.5)
```

```python
X = pd.DataFrame(X)
Y = pd.DataFrame(X_new)
X['src'] = 1 # 1 means source distribution
Y['src'] = 0 # 0 means target distribution
Z = pd.concat([X, Y], ignore_index=True, axis=0)

labels = Z['src'].values
Z = Z.drop('src', axis=1).values
X, Y = X.values, Y.values
```

```python
# clf = RFC(n_estimators=1000, max_depth=100)
clf = MLP(solver='lbfgs', hidden_layer_sizes=(5), random_state=10)
clf.fit(Z, labels)
predictions = clf.predict_proba(Z)[:, 1]

predictions_Z = predictions[:len(X)]
weights = (1./predictions_Z) - 1. 
weights /= np.mean(weights)
plt.figure(figsize = (12,12))
# ax = plt.subplot(211)
# plt.xlabel('Computed sample weight')
# plt.ylabel('# Samples')
# plt.title("Random Foreset")
# sns.distplot(weights, kde=False)

ax = plt.subplot()
Zsize = 0.1 + weights*20
plt.scatter(X[:,0], X[:,1], marker='o', s=Zsize, c='b', label='src', alpha=0.5)
plt.scatter(Y[:,0], Y[:,1], marker='o', s=5, c='g', label='tar')
plt.legend()
```

```python
corr_mat, _ = weighted_corrcoef(X[:,:2], weights)
print(corr_mat)
print(1/np.linalg.norm(weights/size)**2)
```
