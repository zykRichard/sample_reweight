### 1.Load and preprocess data

```python
n, p = 97, 8
n_train, n_test = 67, 30
x_train, y_train = np.zeros((n_train, p)), np.zeros((n_train, 1))
x_test, y_test = np.zeros((n_test, p)), np.zeros((n_test, 1))
# Load data
fd = open('cancer.data', 'r')
data = fd.readlines()
count_train, count_test = 0, 0
for line in data[1:]:
    temp = line.strip().split()
    if temp[-1] == 'T':
        for i in range(1, p+1):
            x_train[count_train, i-1] = float(temp[i])
        y_train[count_train] = float(temp[-2])
        count_train += 1
    else:
        for i in range(1, p+1):
            x_test[count_test, i-1] = float(temp[i])
        y_test[count_test] = float(temp[-2])
        count_test += 1
fd.close()
# Scale
scaler = preprocessing.StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
features = ['lcavol', 'lweight', 'age', 'lbph', 'svi', 'lcp', 'gleason', 'pgg45']
corrmat = np.corrcoef(x_train, rowvar = False)
f, ax = plt.subplots(figsize=(10, 7))
plt.xticks(rotation='90')
sns.heatmap(corrmat, annot=True, square=True, linewidths=.5,
            xticklabels=features, yticklabels=features, cmap='YlGnBu')
plt.show()
```

### 3. Our Method

#### 3.1 Learning Rate
综合优化速度和稳定性1e-4差不多是一个合适的学习率

#### 3.2 Weight Upperbound
weight_upper = 10,  weight_upperbound = 5

```python
data_description = 'prostate cancer data'
weight_l2_list = [1e-1, 1, 10]
#w_opt_list, correlation_list, collinearity_list = [], [], []
for weight_l2 in [30, 50, 100]:
    log_name = 'prostate_l2_%.0e' % (weight_l2)
    w_opt, correlation, collinearity = variable_decorrelation(x = x_train, y = y_train, 
                                                              log_name = log_name, data_description = data_description,
                                                              learning_rate=1e-4, weight_l2 = weight_l2, weight_upper = 0, 
                                                              weight_upperbound = 5, max_iter = 15000, display_iter=300)
    w_opt_list.append(w_opt); correlation_list.append(correlation); collinearity_list.append(collinearity)
```

```python
lasso_our = Lasso()
coef_lasso_our = lasso_our.fit(x_train, y_train, weights = w_opt_list[5])
print(cal_prediction_error(y_test, lasso_our.predict(x_test), 'mse'))
```

```python
ols_our = OLS()
coef_ols_our = ols_our.fit(x_train, y_train, sample_weight = w_opt_list[5].reshape((67, )))
print(cal_prediction_error(y_test, ols_our.predict(x_test), 'mse'))
```
